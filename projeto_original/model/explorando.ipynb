```python
from sklearn.preprocessing import StandardScaler

# Separar as variáveis independentes e a variável alvo
X = data.drop(columns=['Class'])
y = data['Class']

# Normalizar os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Verificar as primeiras linhas dos dados normalizados
print(pd.DataFrame(X_scaled, columns=X.columns).head())
```

```python
# Verificar a presença de valores ausentes no dataset
missing_values = data.isnull().sum()
missing_values = missing_values[missing_values > 0]
print(missing_values)

# Remover as linhas com valores ausentes
data_cleaned = data.dropna()

# Separar as variáveis independentes e a variável alvo
X_cleaned = data_cleaned.drop(columns=['Class'])
y_cleaned = data_cleaned['Class']

# Normalizar os dados
X_scaled_cleaned = scaler.fit_transform(X_cleaned)

# Verificar as primeiras linhas dos dados normalizados
print(pd.DataFrame(X_scaled_cleaned, columns=X_cleaned.columns).head())
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Método do cotovelo para encontrar o número ideal de clusters
sse_cleaned = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled_cleaned)
    sse_cleaned.append(kmeans.inertia_)

# Plotar o gráfico do método do cotovelo
plt.figure(figsize=(10, 6))
plt.plot(k_values, sse_cleaned, marker='o')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Soma dos Erros Quadrados (SSE)')
plt.title('Método do Cotovelo (Dados Limpos)')
plt.grid(True)
plt.show()
```

```python
# Aplicar KMEANS com o número ideal de clusters (definido com base no método do cotovelo)
kmeans = KMeans(n_clusters=4, random_state=42) # Ajuste o número de clusters conforme necessário
clusters_kmeans = kmeans.fit_predict(X_scaled_cleaned)

# Adicionar os clusters ao dataframe original
data_cleaned['Cluster_KMeans'] = clusters_kmeans

# Verificar os resultados
print(data_cleaned.head())
```

```python
from sklearn.cluster import DBSCAN

# Aplicar DBSCAN (ajuste os parâmetros eps e min_samples conforme necessário)
dbscan = DBSCAN(eps=0.5, min_samples=5) # Ajuste os parâmetros conforme necessário
clusters_dbscan = dbscan.fit_predict(X_scaled_cleaned)

# Adicionar os clusters ao dataframe original
data_cleaned['Cluster_DBSCAN'] = clusters_dbscan

# Verificar os resultados
print(data_cleaned.head())
```

```python
import seaborn as sns

# Plotar os clusters gerados pelo KMEANS
plt.figure(figsize=(14, 7))
sns.scatterplot(data=data_cleaned, x='V1', y='V2', hue='Cluster_KMeans', palette='viridis', s=50)
plt.title('Clusters KMEANS')
plt.show()

# Plotar os clusters gerados pelo DBSCAN
plt.figure(figsize=(14, 7))
sns.scatterplot(data=data_cleaned, x='V1', y='V2', hue='Cluster_DBSCAN', palette='viridis', s=50)
plt.title('Clusters DBSCAN')
plt.show()
```
